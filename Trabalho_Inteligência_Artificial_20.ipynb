```python
#valores iniciais para os pesos:
w = [1, 1, 1, 1, 1, 1, 1]
```

```python
#Valores iguais ou maiores que 0 receberão 1, menores que 0 receberão -1
ativacao = 0
```

```python
taxa_aprendizado = 0.01
```

```python
epoca = 500
```

```python
bias = -30
```

```python
for j in range(0, epoca):
  print("Época: ",j,)

  for i in range(0, amostras.shape[0]):

    esperado_a = esperado[i]
    amostra = amostras[i]
    x1 = amostra[0]
    x2 = amostra[1]
    x3 = amostra[2]
    x4 = amostra[3]
    x5 = amostra[4]
    x6 = amostra[5]
    x7 = amostra[6]

    #fazendo a soma dos x com os w, e retornando 1 ou -1 conforme o resultado esperado
    soma = x1*w[0] + x2*w[1] + x3*w[2] + x4*w[3] + x5*w[4] + x6*w[5] + x7*w[6] + bias

    if soma >= 0:
      atual = 1
    else:
      atual = -1

    erro = esperado_a - atual
    
    #atualizando os w:
    w[0] = w[0] + (erro*taxa_aprendizado*x1)
    w[1] = w[1] + (erro*taxa_aprendizado*x2)
    w[2] = w[2] + (erro*taxa_aprendizado*x3)
    w[3] = w[3] + (erro*taxa_aprendizado*x4)
    w[4] = w[4] + (erro*taxa_aprendizado*x5)  
    w[5] = w[5] + (erro*taxa_aprendizado*x6)
    w[6] = w[6] + (erro*taxa_aprendizado*x7)

    print("Predição: ",atual, " Correto: ", esperado_a, " erro: ", erro,)
  print(w)
  print("......................................................")
 
    
```
