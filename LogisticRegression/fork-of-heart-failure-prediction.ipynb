```python
import numpy as np
```

```python
df["serum_creatinine"] = np.log(df["serum_creatinine"])
df['platelets'] = np.sqrt(df['platelets'])
df['creatinine_phosphokinase'] = np.log(df['creatinine_phosphokinase'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
f = ['age', 'anaemia', 'creatinine_phosphokinase', 'diabetes',
       'ejection_fraction', 'high_blood_pressure', 'platelets',
       'serum_creatinine', 'serum_sodium', 'sex', 'smoking', 'time']
l = ['DEATH_EVENT']
```

```python
Y = df[l]
```

```python
X = df[f]
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, shuffle =True,random_state=4)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
f1s = ['age', 'creatinine_phosphokinase', 
       'ejection_fraction',  'platelets',
        'serum_sodium', 'time']
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X_train = sc.fit_transform(X_train[f1s])
X_test = sc.transform(X_test[f1s])
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_test, y_test)
```
