```python
import pandas as pd
```

```python
df = pd.read_csv("/content/data_science_linkedin.csv")
```

```python
df.head(20)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
documents = df['descrição'].values.astype("U")
```

```python
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
my_stop_words = ["a", "o", "em", "para", "que", "com", "as", "os", "você", "dexco", "tu", "cidade", "as", "nas", "nas", "/n", "de", "como", "nossos", "de", "da",
                 "das", "dos", "na", "mais", "no", "se", "está", "localização", "nosso", "do", "uma", "um", "por", "ou", "and", "to", "the", "somos", "seu", "ey",
                 "sua", "todas", "sobre"]
```

```python
vectorizer = TfidfVectorizer(stop_words=my_stop_words)
features = vectorizer.fit_transform(documents)
```

```python
k = 5
model = KMeans(n_clusters=k, init='k-means++', max_iter=100, n_init=1)
model.fit(features)
```

```python
df['cluster'] = model.labels_
```

```python
df[["empresa", "cluster"]]
```

```python
clusters = df.groupby('cluster') 
```

```python
for cluster in clusters.groups:
    f = open('cluster'+str(cluster)+ '.csv', 'w') # create csv file
    data = clusters.get_group(cluster)[['empresa','descrição']] # get title and overview columns
    f.write(data.to_csv(index_label='id')) # set index to id
    f.close()
```

```python
print("Cluster centroids: \n")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
```

```python
for i in range(k):
    print("Cluster %d:" % i)
    for j in order_centroids[i, :10]: #print out 10 feature terms of each cluster
        print (' %s' % terms[j])
    print('------------')
```
